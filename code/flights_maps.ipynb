```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
maps = pd.read_csv('../data/us-airports.csv')
```

```python
maps.head()
```

```python
maps.drop(0, inplace = True)
maps.reset_index(drop = True, inplace = True)
```

```python
maps.head()
```

```python
maps['type'].value_counts()
```

```python
maps.shape
```

```python
maps['country_name'].value_counts()
```

```python
maps = maps[(maps['type'] == 'small_airport') | (maps['type'] == 'medium_airport') | (maps['type'] == 'large_airport')]
```

```python
maps.isnull().sum()
```

```python
maps = maps[~maps['iata_code'].isnull()]
```

```python
maps.isnull().sum()
```

```python
maps.drop(columns = ['id', 'ident', 'elevation_ft', 'continent', 'country_name', 'iso_country', 'region_name', 'iso_region',
                    'local_region', 'municipality', 'scheduled_service', 'gps_code', 'local_code', 'home_link', 'wikipedia_link', 'keywords',
                    'score', 'last_updated'], inplace = True)
maps.reset_index(drop = True, inplace = True)
```

```python
maps.head()
```

```python
maps.shape
```

```python
maps.rename(columns = {'latitude_deg': 'latitude', 'longitude_deg': 'longitude'}, inplace = True)
```

```python
maps.describe()
```

```python
full_set = pd.read_parquet('../data/cleaned_delays_full.parquet')
```

```python
full_set.head()
```

```python
unique_codes = list(full_set['origin'].unique())
```

```python
maps = maps[maps['iata_code'].isin(unique_codes)]
maps.reset_index(drop = True, inplace = True)
```

```python
maps.shape
```

```python
maps.to_csv('../data/airport_lat_long.csv')
```
